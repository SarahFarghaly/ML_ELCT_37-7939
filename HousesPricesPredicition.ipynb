```python
from sklearn.linear_model import LinearRegression 

reg = LinearRegression()
```

```python
labels = house_data['price']
conv_dates = [1 if values == 2014 else 0 for values in house_data.date]
house_data['date'] = conv_dates
train1 = house_data.drop(['id', 'price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(train1, labels, 
                                                       test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_test, y_test)
```

```python
from sklearn import ensemble 
# n_estimators : No. of Boosting stages to perform
# max_depth : Depth of tree node
# learning_rate : Rate of learning Data
# min_samples_split : No. Samples to be splitted for learning data
# loss : Loss Fn. to be optimized (ls --> least squares regression)

clf = ensemble.GradientBoostingRegressor(n_estimators = 400, 
                                         max_depth = 5, min_samples_split =2, 
                                         learning_rate = 0.1, loss = 'ls')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
t_sc = np.zeros((clf.n_estimators), dtype=np.float64)
```

```python
y_pred = reg.predict(x_test)
```

```python
for i,y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i]=clf.loss_(y_test,y_pred)
```

```python
testsc = np.arange(clf.n_estimators)+1
```

```python
# Normalize Data
#from sklearn.datasets import load_iris 
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import Normalizer

#iris = load_iris()
#print(iris.data.shape)

#X = iris.data
#Y = iris.target

norma_tsc = preprocessing.normalize(t_sc)
```

```python
norma_testsc = preprocessing.normalize(testsc)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
plt.plot(testsc,t_sc,'r-',label = 'set dev test')
```
